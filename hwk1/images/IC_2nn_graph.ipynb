```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(plt.style.available)
```

```python
plt.style.use('default')
```

```python
np.arange(10)
```

```python
X = (np.hstack((np.arange(10), [9.2, 9.4, 9.6, 9.8, 10]))) / 10
X
```

```python
Y = np.array([0, 0, 0, 7, 28, 55, 92, 148, 293, 686, 855, 1068, 1341, 1653, 2048]) / 2048
Y
```

```python
plt.plot(X, Y);
plt.axvline(0.8, ymin = 0, ymax = Y[8] + 0.02, ls = '--')
plt.axhline(Y[8], xmin = 0,xmax = X[8], ls = '--')
plt.xlabel(r"nnThreshold ");
plt.ylabel(r"Percentage of kept matches");
```

```python
Ygeo = np.array([0, 0, 0, 6, 27, 53, 81, 104, 123, 151]) / 2048
Ygeo
```

```python
plt.plot(X, Y, label = "2nn");
plt.plot(X[:len(Ygeo)], Ygeo, c = 'r', label = "geo verif");
plt.axvline(0.8, ymin = 0, ymax = Y[8] + 0.02, ls = '--')
plt.axhline(Y[8], xmin = 0,xmax = X[8], ls = '--')
plt.xlabel(r"nnThreshold ");
plt.ylabel(r"Percentage of kept matches");
plt.legend(fontsize = "x-large");
```
